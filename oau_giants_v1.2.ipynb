**Reorder columns in pandas:**
https://datagy.io/reorder-pandas-columns/

```python
def reorder_columns(dataframe, col_name, new_position):
    """Reorder a dataframe's column.

    Args:
        dataframe (pd.DataFrame): dataframe to use
        col_name (string): column name to move
        position (0-indexed position): where to relocate column to

    Returns:
        pd.DataFrame: re-assigned dataframe
    """
    temp_col = dataframe[col_name]
    dataframe = dataframe.drop(columns=[col_name])
    dataframe.insert(loc=new_position, column=col_name, value=temp_col)

    return dataframe
```

```python
# reordering columns 
giants_copy = reorder_columns(dataframe=giants_copy, col_name='Time24h', new_position=1)
giants_copy = reorder_columns(dataframe=giants_copy, col_name='Venue', new_position=2)
giants_copy = reorder_columns(dataframe=giants_copy, col_name='CompGroup', new_position=3)
giants_copy = reorder_columns(dataframe=giants_copy, col_name='CompID', new_position=4)
giants_copy = reorder_columns(dataframe=giants_copy, col_name='MatchID', new_position=8)
giants_copy = reorder_columns(dataframe=giants_copy, col_name='FORM', new_position=15)
giants_copy = reorder_columns(dataframe=giants_copy, col_name='FormPen', new_position=18)

giants_copy.head(1)
```

# CORRECT FORM AND PEN FORM
Programmatically inspect the Win(W), Draw(D) or Loss(L) for:
- regular match time result **(FTR)** and
- penalty shoot-out! **(PSHTR)**

```python
#Regular time results
conditions = [
    #When OAU Giants are the HomeTeam
    (giants_copy['FTHG'] < giants_copy['FTAG']) & (giants_copy['HomeTeam']=='OAU Giants'),
    (giants_copy['FTHG'] == giants_copy['FTAG']) & (giants_copy['HomeTeam']=='OAU Giants'),
    (giants_copy['FTHG'] > giants_copy['FTAG']) & (giants_copy['HomeTeam']=='OAU Giants'),
    
    #When OAU Giants are the AwayTeam
    (giants_copy['FTHG'] > giants_copy['FTAG']) & (giants_copy['AwayTeam']=='OAU Giants'),
    (giants_copy['FTHG'] == giants_copy['FTAG']) & (giants_copy['AwayTeam']=='OAU Giants'),
    (giants_copy['FTHG'] < giants_copy['FTAG']) & (giants_copy['AwayTeam']=='OAU Giants'),
    ]

values = ['L', 'D', 'W', 'L', 'D', 'W']

giants_copy['FTR'] = np.select(conditions, values)
giants_copy.head(10)
```

```python
#Penalty shoot-out results
conditions = [
    #When OAU Giants are the HomeTeam
    (giants_copy['PENHG'] < giants_copy['PENAG']) & (giants_copy['HomeTeam']=='OAU Giants'),
    (giants_copy['PENHG'] > giants_copy['PENAG']) & (giants_copy['HomeTeam']=='OAU Giants'),
        
    #When OAU Giants are the AwayTeam
    (giants_copy['PENHG'] < giants_copy['PENAG']) & (giants_copy['AwayTeam']=='OAU Giants'),
    (giants_copy['PENHG'] > giants_copy['PENAG']) & (giants_copy['AwayTeam']=='OAU Giants'),
    ]

values = ['L', 'W', 'W', 'L']

giants_copy['PSHTR'] = np.select(conditions, values, np.nan)
giants_copy.head(10)
```

```python
def Form_Waffle(period):
    # Create Figure & Plot Axes
    fig, ax = plt.subplots(1,1, figsize=(12, 4), facecolor=facecolor, layout='constrained')
    ax.set_facecolor(facecolor)
    ax.axis('off')

    # plot variables
    pad_y = .18
    pad_x = .06
    st_x = .1     #starting x-pos
    st_y = 0     #starting y-pos

    #  Check count of df or list or array to know how many boxes are needed
    column = str(period)
    waffle_data = results_df[f'{column}'].to_list()
    num_of_boxes = len(waffle_data) 

    color_dict = {'W' : '#005A32', #E0CB8A
                  'D' : '#d6d37f', ##79A19D #C3B59F
                  'L' : 'grey'}

    # Making the Waffle Plot
    for i in range(num_of_boxes): #total num of boxes to be plotted
        #SKnow When to switch to the next column
        if i % 5 == 0 and i > 0:  
            st_x += pad_x
            st_y = 0

        #Default Waffle Plot
        waffle_box = plt.Rectangle((st_x, st_y), .05, .15,
                                   fc=color_dict[waffle_data[i]], ec='none') 
        st_y += pad_y    

        # Add to Axes    
        ax.add_patch(waffle_box)

    # Title Text
    title = f"{team} FORM GUIDE ({column}) <WINS>, <DRAWS> AND <LOSSES> IN ALL COMPS {data_span}"
    # Set titles for the figure and the subplot respectively
    h_fig(x=.12, y=1.18, s=title, color=off_white, highlight_textprops=[{'color':'#005A32'},
                                                                        {'color':'#d6d37f'},
                                                                        {'color':'grey'}],
       font=t_font, fontsize=t_fsize, fontweight='bold', zorder=2)

    # Endnotes
    endnote_text = ("FTR — Fulltime Results  |   HTR — Halftime Results  |   SHR — Second Half Results\n"
                    "Matches are arranged from Bottom to Top in each column\n"
                   "(Walkover & Abandoned ties are excluded)")
    # writing endnote text on figure
    endnote = fig.text(x=.51, y= -.23, s=endnote_text, color=off_white, linespacing= 2,
                   ha='center', va='center', font=b_font, fontsize=b_fsize, zorder=2)

    # Name Text
    name_text = fig.text(x=0.51, y= -.45, s=author, ha='center', va='center', 
                   font=b_font, color=plot_color, alpha=.3, fontsize=b_fsize, zorder=2)

    # Path Effects for texts
    import matplotlib.patheffects as path_effects
    endnote.set_path_effects([path_effects.Stroke(linewidth=.018, foreground=off_white),
                          path_effects.Normal()])
    name_text.set_path_effects([path_effects.Stroke(linewidth=.005, foreground=plot_color), 
                            path_effects.Normal()])

    # Figure Paddings         
    pad_top = "\n\n"
    pad_end = "\n\n"

    fig.text(x=0.19, y= 1.20, s=pad_top, color=off_white, linespacing= 2,
                ha='center', va='center', font=b_font, fontsize=b_fsize, zorder=2)

    fig.text(x=0.51, y= -.50, s=pad_end, color=off_white, linespacing= 2,
                ha='center', va='center', font=b_font, fontsize=b_fsize, zorder=2)

    # Save and display
#     plt.savefig(f'{period}_{team}_FORM_GUIDE.png', dpi=1000, bbox_inches = "tight")
    plt.show();
```

```python
# Define a function to calculate streaks
def calculate_streaks(result_column, streak_type):
    streaks = []
    current_streak = 0

    for result in result_column:
        if result == f'{streak_type}':    #Replace With D or L
            current_streak += 1
        else:
            streaks.append(current_streak)
            current_streak = 0

    # Append the last streak
    streaks.append(current_streak)

    return streaks
```
